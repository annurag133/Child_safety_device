```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
import pickle 
```

```python
df=pd.read_csv('temp.csv')
df.head(5)
```

```python
df.drop("Date",axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
#we are going to use 2 models for Tmin And Tmax
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5), dpi=200)
plt.plot(df["Next_Tmax"], label="Next_Tmax")
plt.plot(df["Next_Tmin"],label = "Next_Tmin")
plt.legend()
plt.show()
```

```python
cor = df.corr()
plt.figure(figsize=(15,7),dpi=100)
sns.heatmap(cor)
plt.show()
```

```python
df = df.dropna()
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
train,test = train_test_split(df,test_size=0.2,random_state=0)
```

```python
test.head()
```

```python
def Xy(df):
    X = df.drop(["Next_Tmax","Next_Tmin"],axis=1)
    y_max = df["Next_Tmax"]
    y_min = df["Next_Tmin"]
    
    return X,y_max,y_min
```

```python
X_train, y_max_train, y_min_train = Xy(train)
X_test, y_max_test, y_min_test = Xy(test)
```

```python
y_min_train.head()
```

```python
max_model = LinearRegression()
max_model.fit(X_train, y_max_train)
```

```python
y_max_pred = max_model.predict(X_test)
max_model.score(X_test,y_max_test) # Accuracy
```

```python
min_model = LinearRegression()
min_model.fit(X_train, y_min_train)
```

```python
y_min_pred = min_model.predict(X_test)
min_model.score(X_test,y_min_test) # Accuracy
```

```python
pickle.dump(max_model, open('T_max.pkl','wb'))
pickle.dump(min_model, open('T_min.pkl', 'wb'))
```
